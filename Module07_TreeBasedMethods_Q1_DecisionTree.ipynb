## Using the decision tree algorithm

```python
# Separating the independent and dependent variables
X = file[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
Y = file['Species']
#X
#Y
```

#### Training the data and showing the values of Sensitivity, Specificity, Precision, Recall,and F1 score for different test sizes

```python
splits = [0.25,0.30,0.35,0.40] #The different test size ratios
accu = [] #An empty list that the accuracy values can be appended onto

# Creating the classifier
clf = DecisionTreeClassifier(criterion='gini', max_depth=3)

for testsize in splits:
    # Train and test splitting
    X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = testsize)

    # Training 
    clf.fit(X_train,y_train)

    # Creating a series with predictions
    y_pred = clf.predict(X_test)

    accu.append(metrics.accuracy_score(y_test, y_pred))
    
    print('For test size =',testsize,':','\n')
    print('The Precision, Recall, and F1 score:')
    print('________________________________________________________\n\n'
          ,classification_report(y_test, y_pred),
          '\n________________________________________________________\n\n')
    
    print('The confusion matrix: \n')#,
#           '\n- - - - - - -\n',
#           confusion_matrix(y_test, y_pred, labels = ['Iris-setosa','Iris-versicolor','Iris-virginica']),
#           '\n- - - - - - -\n\n')

    # Heatmap showing the confusion matrix values
    classnames = ['Iris-setosa','Iris-versicolor','Iris-virginica']
    classnames_short = ['setosa','versicolor','virginica']
    sns.set(rc = {'figure.figsize':(3,2.5)})
    mx = confusion_matrix(y_test, y_pred, labels = classnames)
    sns.heatmap(mx, annot = True, xticklabels = classnames_short, yticklabels = classnames_short)
    plt.show()
    print('\n')
    
    # Calculating sensitivity 
    print('Sensitivity\n______________________________________\n')
    for i in range(len(classnames)):
        tp = mx[i][i]                                             # tp is true positive
        fn = sum(mx[i])                                           # fn is false negative
        print ('For',classnames[i],', sensitivity =',
                round(tp/fn,2))                      
    print('______________________________________\n\n')
    
    # Calculating specificity
    print('Specificity\n______________________________________\n')
    for i in range(len(classnames)):
        tn = sum(sum(mx)) - sum(mx)[i] - sum(mx[i]) + mx[i][i]    #tn is true negative
        fp = sum(mx.transpose()[i]) - mx[i][i]                    #fp is false positive
        print ('For',classnames[i],', specificity =',
                round(tn/(tn+fp),2))        
    print('______________________________________\n**************************************\n')
```

### For AUC and ROC 

#### The ROC curves can only be plotted when it is a binary classification problem. 
#### Here, to plot those curves, we use one hot encoding on the 'Species' column and plot the curves for each class separately.

```python
# Separating the 'Species' column into 3 different columns with binary values, one for each species, and saving it in a new dataframe.
file2 = pd.get_dummies(file)
file2.head()
```

#### Training the data and displaying ROC curves

```python
for testsize in splits:
    
    for i in range(len(['Species_Iris-setosa','Species_Iris-versicolor','Species_Iris-virginica'])):  
        # Separating the independent and dependent variables
        X2 = file2[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
        Y2 = file2[['Species_Iris-setosa','Species_Iris-versicolor','Species_Iris-virginica'][i]]

        # Creating the classifier
        clf = DecisionTreeClassifier(criterion='gini', max_depth=3)

        # Train and test splitting
        X2_train, X2_test, y2_train, y2_test = train_test_split(X2, Y2, test_size = testsize)

        # Training 
        clf.fit(X2_train,y2_train)

        #define metrics
        y2_pred_proba = clf.predict_proba(X2_test)[::,1]
        fpr, tpr, _ = roc_curve(y2_test,  y2_pred_proba) #fpr: false positive rate; tpr: true positive rate
        auc = round(roc_auc_score(y2_test, y2_pred_proba),3)

        #create ROC curve
        from math import *
        c1 = (2**i)-2*i
        c2 = (cos(i*pi) - 1)/-2
        c3 = (2**(abs(i-2)))-2*(abs(i-2))
        rnclr = [c1, c2, c3]
         
        sns.set(rc = {'figure.figsize':(5.5,5.5)})
        plt.plot(fpr,tpr,label="AUC="+str(auc),color = rnclr)
        plt.ylabel('True Positive Rate')
        plt.xlabel('False Positive Rate')
        plt.title('ROC curve for testsize = ' + str(testsize)+'\n', fontsize = '15')
        first_legend = plt.legend(('setosa', 'versicolor', 'virginica'), loc = 8)
        plt.gca().add_artist(first_legend)
        plt.legend(loc = 'lower right')
        
    plt.show()  
    print ('\n__________________________________________________')
```
